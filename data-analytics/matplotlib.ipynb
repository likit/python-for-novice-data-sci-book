## **Matplotlib Plot Types**


*   Line Plots
*   Bar Plots
*   Scatter Plots
*   Histogram Plots (Distribution)



```python
import matplotlib.pyplot as plt
import numpy as np
```



> **Line Plots**



```python
# Line Plots
x = np.arange(20)
y = np.random.rand(20)

plt.plot(x,y)
```



> **Basic Plot Parameters**

*   Line Color https://matplotlib.org/stable/gallery/color/named_colors.html
*   Marker types https://matplotlib.org/stable/api/markers_api.html
*   Line style https://matplotlib.org/3.5.1/gallery/lines_bars_and_markers/linestyles.html






```python
# Line Plot Parameters
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)

# Plot Labeling
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('TITLE TEXT')
```

> **Bar Plots**

```python
# Bar Plots
x = np.arange(20)
y = np.random.rand(20)

plt.bar(x,y)
```

> **Scatter Plots**

```python
# Scatter Plots
x = np.arange(20)
y = np.random.rand(20)

plt.scatter(x,y)
```

> **Histogram Plots**

*   Plotting Data Distribution




```python
# Histogram Plots or Distribution Plots
y = np.random.randn(100) #normal dist random
plt.hist(y)

# Labeling
plt.xlabel('Random Values')
plt.ylabel('Density')
plt.title('Random Value Distribution using np.random.randn()')
```

> **Box Plots**

*   Another Data Distribution Plots
*   https://www.tutorialspoint.com/matplotlib/matplotlib_box_plot.htm
*   https://matplotlib.org/3.1.1/gallery/pyplots/boxplot_demo_pyplot.html#sphx-glr-gallery-pyplots-boxplot-demo-pyplot-py



```python
# np.random.normal(mean, std, number-of-values) -> generate normal distributed array
# matplotlib.pyploy.boxplot(dataArray)

data1 = np.random.normal(100, 10, 200)
bp = plt.boxplot(data1)

##     min   Q1 median Q3   max  outliers
##oo   |-----|-----|     o
```

```python
# Multiple Samples/Arrays

data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
dataToPlot = [data1, data2, data3]
# print(dataToPlot)

bp = plt.boxplot(dataToPlot)
```

> **Violin Plots**

*   Similar to box plot
*   Show the probability density of the data

```python
# Violin Plot
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
dataToPlot = [data1, data2, data3]

vp = plt.violinplot(dataToPlot)
```

## **Multiple Plots in 1 Axis**

*   Add Plot Legends



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)

y1 = np.cos(x)
y2 = (x-np.mean(x))/4 #random calculation
y3 = np.random.randn(len(x)) #same size as X

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
# plt.plot(x, y1, color='darkorchid')
# plt.plot(x, y2, color='deepskyblue')
# plt.plot(x, y3, color='salmon')

# labeling and legend
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(['y1', 'y2', 'y3'], loc='upper right')
```

## **Subplots**

*   Set of subplots (multiple axes) in one figure



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)

y1 = np.cos(x)
y2 = (x-np.mean(x))/4 #random calculation
y3 = np.random.randn(len(x)) #same size as X
```

```python
# Subplots(columns, rows)
fig, ax = plt.subplots(1,3)

ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)
```

```python
# Subplots(columns, rows)
fig, ax = plt.subplots(1,3, figsize=[16,5]) #figsize = [width, height]

# Plotting
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)

# Labeling
ax[0].set_ylabel('Y Axis Value')
ax[0].set_xlabel('Plot #1 SinWave')
ax[1].set_xlabel('Plot #2 Linear')
ax[2].set_xlabel('Plot #3 Random')
fig.suptitle('Global Plot Title')
```

```python
# Compare Violinplot to Boxplot
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
dataToPlot = [data1, data2, data3]

fig, ax = plt.subplots(1,2, figsize=[9,4])
vp = ax[0].violinplot(dataToPlot)
bp = ax[1].boxplot(dataToPlot)
```
